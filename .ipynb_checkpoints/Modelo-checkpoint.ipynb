# Proyecto de clasificación de mariposas

## Entrenar el modelo

```python
# Importamos las librerías
import sys
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import optimizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Activation

#Almacenamos el directorio donde tenemos las imagenes para entrenar
data_entrenamiento ='./Dataset Mariposas/Entrenamiento'
data_pruebas ='./Dataset Mariposas/Pruebas'

# Parametros de la red neuronal
epocas = 20 #Número de veces que vamos a iterar sobre nuestro set de datos de entrenamiento
altura, longitud = 400, 400 #Tamaño al que vamos a procesar las imagenes
batch_size = 1 #Número de imagenes que el modelo procesará antes de actualizar los pesos de la red neuronal durante el entrenamiento
pasos = 1000 #Número de veces que se va a procesar la información en cada una de las epocas
pasos_pruebas = 200 #Lo mismo de arriba pero con nuestros datos de prueba
filtrosConv1 = 32 #Número de filtros que hagamaos en cada convolución
filtrosConv2 = 64
tamano_filtro1 = (3,3)
tamano_filtro2 = (2,2)
tamano_pool = (2,2) #Tamaño del filtro usado en el Max Pooling
clases = 10
lr = 0.0005 #Ajustes que hace la red neuronal para acercarse a una solución optima

# Preprocesamiento de las imagenes
entrenamiento_datagen = ImageDataGenerator(
    rescale = 1./255, # Rescala las imagenes
    shear_range = 0.3, # Inclina las imagenes
    zoom_range = 0.3, # Le hace zoom a las imagenes
    horizontal_flip = True # Invierte las imagenes
)

pruebas_datagen = ImageDataGenerator(
    rescale = 1./255
)

imagen_entrenamiento = entrenamiento_datagen.flow_from_directory(
    data_entrenamiento,
    target_size = (altura, longitud),
    batch_size = batch_size,
    class_mode = 'categorical'
)

imagen_pruebas = pruebas_datagen.flow_from_directory(
    data_pruebas,
    target_size = (altura, longitud),
    batch_size = batch_size,
    class_mode = 'categorical'
)

# Crear la Red Convolucional
cnn = Sequential()

# Añadimos la primera capa de convolución
cnn.add(Conv2D(filtrosConv1, tamano_filtro1, padding = 'same', input_shape = (altura, longitud, 3), activation = 'relu'))

# Añadimos una capa de Max Pooling
cnn.add(MaxPooling2D(pool_size = tamano_pool))

# Añadimos la segunda capa de convolución
cnn.add(Conv2D(filtrosConv2, tamano_filtro2, padding = 'same', activation = 'relu'))

# Añadimos una segunda capa de Max Pooling
cnn.add(MaxPooling2D(pool_size = tamano_pool))

# Empezamos la clasificación
cnn.add(Flatten())

cnn.add(Dense(256, activation = 'relu'))

cnn.add(Dropout(0.5))

cnn.add(Dense(clases, activation = 'softmax'))

cnn.compile(loss ='categorical_crossentropy', optimizer=optimizers.Adam(learning_rate=lr), metrics = ['accuracy'])

cnn.fit(imagen_entrenamiento, steps_per_epoch = pasos, epochs = epocas, validation_data = imagen_pruebas, validation_steps = pasos_pruebas)

dir = './modelo/'

if not os.path.exists(dir):
    os.mkdir(dir)
cnn.save('./modelo/modelo.h5')
cnn.save_weights('./modelo/pesos.h5')
```
